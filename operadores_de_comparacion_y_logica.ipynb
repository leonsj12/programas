### **Introducción**
En este cuaderno exploraremos las comparaciones numéricas en Python. Las comparaciones son fundamentales en la programación, ya que permiten la toma de decisiones en base a condiciones lógicas.

Python proporciona operadores de comparación como `>`, `<`, `==`, `!=`, `and` y `or`, los cuales permiten evaluar relaciones entre valores y determinar si una condición es verdadera (`True`) o falsa (`False`).

### Objetivos
1. Comprender el uso de los operadores de comparación en Python.
2. Evaluar expresiones lógicas utilizando operadores relacionales (`>`, `<`, `==`, `!=`).
3. Aplicar operadores lógicos (`and`, `or`) para combinar múltiples condiciones.
4. Utilizar `pandas` para organizar y visualizar comparaciones en forma de tabla.

```python
# 1. Comparar si un número es mayor que otro
print(5 > 3)  # True

# 2. Comparar si un número es menor que otro
print(5 < 8)  # True

# 3. Comparar si dos números son iguales
print(7 == 8)  # False

# 4. Comparar si dos números son diferentes
print(7 != 8)  # True

# 5. Comparación lógica con AND y OR
print(5 < 3 and 8 < 5)  # False
print(3 != 5 or 8 == 8)  # True

```

```python
import pandas as pd

# Crear un diccionario con las comparaciones y sus resultados
data = {
    "Comparación": [
        "5 > 3",
        "5 < 8",
        "5 >= 8",
        "8 <= 5",
        "7 == 8",
        "7 != 8",
        "5 < 3 and 8 < 5",
        "3 != 5 or 8 == 8"
    ],
    "Resultado Booleano": [
        5 > 3,
        5 < 8,
        5 >= 8,
        8 <= 5,
        7 == 8,
        7 != 8,
        (5 < 3 and 8 < 5),
        (3 != 5 or 8 == 8)
    ]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Mostrar la tabla
df
```

### Conclusiones
- Los operadores de comparación permiten evaluar relaciones entre valores numéricos y devuelven un resultado booleano (`True` o `False`).
- Los operadores lógicos `and` y `or` permiten combinar múltiples condiciones, lo que es útil para la toma de decisiones en programas.
- `pandas` facilita la organización y visualización de datos en forma de tabla, permitiendo una mejor interpretación de los resultados.
- Comprender y utilizar correctamente estas comparaciones es esencial para escribir código más eficiente y efectivo.
